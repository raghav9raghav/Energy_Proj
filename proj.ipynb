```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
df = pd.read_csv(r"C:\Users\Sparsh Sopory\Desktop\europe_energy_sustainability.csv")

country_col = "country"
year_col = "year"

# Clean country names
df[country_col] = df[country_col].str.strip().str.lower()

# Handle missing values
df = df.fillna(df.mean(numeric_only=True))

# Sort data
df = df.sort_values(by=[country_col, year_col])

# Normalize year
df["year_norm"] = df[year_col] - df[year_col].min()

# ---------------------------
# 🔥 PART 1: TREND MODEL (Population + GDP)
# ---------------------------
trend_features = ["population", "gdp"]
trend_models = {}

for feature in trend_features:
    model = LinearRegression()
    
    X_trend = df[[country_col, "year_norm"]]
    Y_trend = df[feature]
    
    X_encoded = pd.get_dummies(X_trend, columns=[country_col])
    
    model.fit(X_encoded, Y_trend)
    trend_models[feature] = (model, X_encoded.columns)

# ---------------------------
# 🔥 PART 2: RANDOM FOREST (Other features)
# ---------------------------
other_features = [col for col in df.columns if col not in trend_features + [country_col, year_col, "year_norm"]]

# Create lag + rolling features
for col in other_features:
    df[f"{col}_lag1"] = df.groupby(country_col)[col].shift(1)
    df[f"{col}_lag2"] = df.groupby(country_col)[col].shift(2)
    df[f"{col}_roll3"] = df.groupby(country_col)[col].rolling(3).mean().reset_index(0, drop=True)

df = df.dropna()

# Features & Target
feature_cols = (
    [country_col, year_col] +
    [f"{col}_lag1" for col in other_features] +
    [f"{col}_lag2" for col in other_features] +
    [f"{col}_roll3" for col in other_features]
)

X = df[feature_cols]
Y = df[other_features]

# Encoding
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown='ignore'), [country_col])
    ],
    remainder='passthrough'
)

# Stronger model
rf_model = RandomForestRegressor(
    n_estimators=700,
    max_depth=20,
    min_samples_split=5,
    random_state=42
)

rf_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("scaler", StandardScaler()),
    ("model", rf_model)
])

# ---------------------------
# 🔥 TRAIN / TEST SPLIT
# ---------------------------
train = df[df[year_col] <= 2019]
test = df[(df[year_col] >= 2020) & (df[year_col] <= 2025)]

rf_pipeline.fit(train[feature_cols], train[other_features])

# Predict RF features
rf_pred = rf_pipeline.predict(test[feature_cols])
rf_pred_df = pd.DataFrame(rf_pred, columns=other_features)

# ---------------------------
# 🔥 TREND PREDICTIONS
# ---------------------------
trend_pred = {}

for feature in trend_features:
    model, cols = trend_models[feature]
    
    X_test_trend = test[[country_col, "year_norm"]]
    X_test_encoded = pd.get_dummies(X_test_trend, columns=[country_col])
    
    X_test_encoded = X_test_encoded.reindex(columns=cols, fill_value=0)
    
    trend_pred[feature] = model.predict(X_test_encoded)

trend_pred_df = pd.DataFrame(trend_pred)

# ---------------------------
# 🔥 FINAL COMBINE
# ---------------------------
final_pred = pd.concat([trend_pred_df, rf_pred_df], axis=1)
actual = test[trend_features + other_features].reset_index(drop=True)

# ---------------------------
# 🔥 EVALUATION
# ---------------------------
mse = mean_squared_error(actual, final_pred)
rmse = np.sqrt(mse)
r2 = r2_score(actual, final_pred)

print("\n📊 FINAL MODEL PERFORMANCE")
print("✅ MSE:", round(mse, 3))
print("✅ RMSE:", round(rmse, 3))
print("✅ R2 Score:", round(r2, 3))

# ---------------------------
# 🔥 COMPARISON OUTPUT
# ---------------------------
comparison = pd.concat(
    [final_pred.add_suffix("_pred"), actual.add_suffix("_actual")],
    axis=1
)

print("\n📊 Sample Comparison:")
print(comparison.head())
```
